```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("vikuiti_reflectance.csv", sep=", ")

# converter nm -> eV
df["E"] = 1240.0 / df["x"]
df = df.sort_values("E", ascending=True)

plt.figure(figsize=(6,4))
plt.plot(df["E"], df["y"], marker="o", linestyle="-")
plt.xlabel("Energia [eV]")
plt.ylabel("y (adimensional)")
plt.title("y vs Energia")
plt.grid(True)
plt.show()

df_renamed = df.rename(columns={"y": "r"})
df_renamed[["E", "r"]].to_json("../configuration/vikuiti_reflectance.json", orient="records", indent=2)
```
